## Acessando MySQL

```python
import mysql.connector
from mysql.connector import Error

try:
    # Conexão ao banco de dados
    conexao = mysql.connector.connect(
        host='localhost',      
        database='ClientesMySQL',  
        user='root',    
        password='1234'   
    )

    if conexao.is_connected():
        print("Conexão bem-sucedida ao MySQL!")
            
        # Executando uma consulta
        cursor = conexao.cursor()
        cursor.execute("SELECT * FROM Clientes")
        resultados = cursor.fetchall()
            
        # Exibindo os resultados
        print("Resultados da consulta:")
        for linha in resultados:
            print(linha)

except Error as e:
    print(f"Erro ao conectar ao MySQL: {e}")
finally:
    # Fechando a conexão
    if 'conexao' in locals() and conexao.is_connected():
        cursor.close()
        conexao.close()
        print("Conexão ao MySQL encerrada.")
```

# Juntar as tabelas SQL Server e My SQL

### DataFrame SQLServer

```python
# DataFrame SQL Server
import pyodbc
import pandas as pd

conn = pyodbc.connect('DRIVER={ODBC Driver 18 for SQL Server};'
                      'SERVER=localhost,1434;'
                      'DATABASE=FinanceiroSQLServer;'
                      'UID=sa;'
                      'PWD=SuaSenhaForteAqui!123;'
                      'Encrypt=yes;TrustServerCertificate=yes;',
                       autocommit=False  
)

# Criar DataFrame diretamente da consulta
df_transacoes = pd.read_sql("SELECT ID_Transacao, Valor, Data_Transacao, ID_Cliente, Categoria_Transacao, Meio_Pagamento FROM FinanceiroSQLServer.dbo.TransacoesFinanceiras", conn)

print(f"DataFrame criado com {len(df_transacoes)} registros")
display(df_transacoes)

conn.close()
```

### DataFrame MySql

```python
# Data Frame My SQL
import mysql.connector
import pandas as pd

try:
    # Conexão ao banco de dados
    conexao = mysql.connector.connect(
        host='localhost',      
        database='ClientesMySQL',  
        user='root',    
        password='1234'   
    )

    if conexao.is_connected():
        print("Conexão bem-sucedida ao MySQL!")
        
        # Criar DataFrame diretamente da consulta
        df_clientes = pd.read_sql("SELECT * FROM Clientes", conexao)
        
        print(f"DataFrame criado com {len(df_clientes)} registros")
        print(df_clientes)

except mysql.connector.Error as e:
    print(f"Erro ao conectar ao MySQL: {e}")
finally:
    # Fechando a conexão
    if 'conexao' in locals() and conexao.is_connected():
        conexao.close()
        print("Conexão ao MySQL encerrada.")
```

### Merge dos DataFrames

```python
# Merge entre transações e clientes
df_merged = pd.merge(df_transacoes, df_clientes, left_on='ID_Cliente', right_on='ID_Cliente', how='inner')
print(f"DataFrame merged com {len(df_merged)} registros")
display(df_merged)
```

## Diagnóstico do Merge

```python
# IDs únicos em cada DataFrame
ids_transacoes = set(df_transacoes['ID_Cliente'].unique())
ids_clientes = set(df_clientes['ID_Cliente'].unique())

print(f"IDs únicos em transações: {len(ids_transacoes)}")
print(f"IDs únicos em clientes: {len(ids_clientes)}")

# IDs que estão em transações mas não em clientes
ids_sem_cliente = ids_transacoes - ids_clientes
print(f"IDs sem cliente: {sorted(ids_sem_cliente)}")

# Quantas transações têm esses IDs
transacoes_sem_cliente = df_transacoes[df_transacoes['ID_Cliente'].isin(ids_sem_cliente)]
print(f"Transações perdidas: {len(transacoes_sem_cliente)}")
print(transacoes_sem_cliente[['ID_Transacao', 'ID_Cliente', 'Valor']])
```

## Df_Agrupado e salvar arquivo csv no Hadoop 

### Df_Agrupado

```python
# DF Agrupado por Ano - Mes - Categoria - Valor Total
df_merged['Ano'] = df_merged['Data_Transacao'].dt.year
df_merged['Mes'] = df_merged['Data_Transacao'].dt.month

df_agrupado = df_merged.groupby(['Ano', 'Mes', 'Categoria_Transacao'])['Valor'].sum().reset_index()
print(f"DataFrame agrupado com {len(df_agrupado)} registros")
display(df_agrupado)
```

```python
# Salvando DataFrame agrupado em CSV na pasta arquivo_csv
df_agrupado.to_csv('arquivo_csv/Transacoes_Ano_Mes_Categoria2.csv', index=False)
df_agrupado.to_csv("Transacoes_Ano_Mes_Categoria.csv", index=False)
print("DataFrame agrupado e salvo como: Transacoes_Ano_Mes_Categoria.csv")
```

```python
from hdfs import InsecureClient
import os

# Caminho local do arquivo
local_folder = "/home/henriquenunes/curso_python/Aula16/arquivo_csv"
file_name = "Transacoes_Ano_Mes_Categoria.csv"
local_file_path = os.path.join(local_folder, file_name)

# Conexão com Hadoop via Docker
hdfs_url = "http://localhost:9870"  
hdfs_user = "root"
hdfs_dest_path = "/Financeiro"

# Criar cliente HDFS
client = InsecureClient(hdfs_url, user=hdfs_user)

# Garantir que a pasta exista no HDFS
client.makedirs(hdfs_dest_path)

# Caminho final no HDFS
hdfs_file_path = f"{hdfs_dest_path}/{file_name}"

# Upload
client.upload(hdfs_file_path, local_file_path, overwrite=True)

print(f"✅ Arquivo enviado com sucesso para: {hdfs_file_path}")
```
